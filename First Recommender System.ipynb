```python
# List to store transaction data
transactions = []

# Generate transaction data
for i in range(1000):
    product_id = random.randint(1, len(products))
    customer_id = random.randint(1, 1000)
    purchase_date = fake.date_between(start_date='-2y', end_date='today').strftime('%Y-%m-%d')
    quantity = random.randint(1, 5)
    sales_channel = random.choice(["Online", "In-Store"])
    rating = round(random.uniform(0.5, 5.0), 1)
    
    transactions.append([product_id, customer_id, purchase_date, quantity, sales_channel, rating])

# Create Transactional DataFrame
transaction_df = pd.DataFrame(transactions, columns=['Item ID', 'Customer ID', 'Purchase Date', 'Quantity', 'Sales Channel', 'Rating'])
```

## Recomender System

#### Calculate Average Rating for Each Product

```python
average_ratings = transaction_df.groupby('Item ID')['Rating'].mean().reset_index()
```

```python
average_ratings.head(10)
```

#### Merge Average Ratings with Inventory Data

```python
product_recommendations = pd.merge(inventory_df, average_ratings, on='Item ID', how='left')
```

#### Recommend Products Based on Highest Average Rating

```python
def recommend_products(num_recommendations=5):
    recommended_products = product_recommendations.sort_values(by='Rating', ascending=False).head(num_recommendations)
    return recommended_products

# Get the top 3 recommended products
top_recommendations = recommend_products(num_recommendations=3)
```
